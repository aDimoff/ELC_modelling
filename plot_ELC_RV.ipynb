```python
import matplotlib.pyplot as plt
import numpy as np
#import join
```

```python
# reset defalult plotting values
plt.rcParams['figure.figsize'] = (16, 7)
plt.rc('font', family='sans-serif')
plt.rc('axes', labelsize=20)
plt.rc('axes', labelweight='bold')
plt.rc('axes', titlesize=30)
plt.rc('axes', titleweight='bold')
plt.rc('axes', linewidth=2)
plt.rc('xtick',labelsize=20,direction='in',top=True)
plt.rc('ytick',labelsize=20,direction='in',right=True)
```

```python
## read input directory name from command line
input_directory = 'HD116514/'
input_RVs_file = input_directory+'HD116514_RVs.dat'
input_fold_file = input_directory+'ELCdataRV1.fold'
input_model_file = input_directory+'star1.RV'
input_model_del_file = input_directory+'star1.delRV'
```

```python
Names = ['time','RV','RV_err']
RV_in = np.genfromtxt(input_RVs_file,names=Names,unpack=False)
Names = ['model_time','model_RV']
model_in = np.genfromtxt(input_model_file,names=Names,unpack=False)
Names = ['fold_time','fold_RV','fold_err']
fold_in = np.genfromtxt(input_fold_file,names=Names,unpack=False)
```

```python
plt.errorbar(RV_in['time'],RV_in['RV'],yerr=RV_in['RV_err'],marker='.',ls='',capsize=4,color='red')
plt.grid()
#plt.title('V530Lyr Radial Velocity')
plt.xlabel('Time (HJD - 2400000)')
plt.ylabel('RV (km/s)')
#plt.savefig('V530Lyrfull.png',bbox_inches=None,dpi=300)
```

```python
plt.plot(model_in['model_time'],model_in['model_RV'],ls='-',lw=2,color='black')
plt.errorbar(fold_in['fold_time'],fold_in['fold_RV'],fold_in['fold_err'],marker='.',ls='',capsize=4,color='red')
plt.errorbar(fold_in['fold_time']+1,fold_in['fold_RV'],fold_in['fold_err'],marker='.',ls='',capsize=4,color='red')
plt.xlim(0,2)
plt.grid()
#plt.title('HD116514 Phased RV')
plt.xlabel('Orbital Phase')
plt.ylabel('RV (km/s)')
plt.savefig('HD116514_phasefold.png',bbox_inches='tight',dpi=300)
```
