```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
%matplotlib inline
plt.rcParams["figure.figsize"] = (18,12)
plt.rc('font', family='sans-serif')
plt.rc('axes', labelsize=22)
plt.rc('axes', labelweight='bold')
plt.rc('axes', titlesize=24)
plt.rc('axes', titleweight='bold')
plt.rc('axes', linewidth=2)
plt.rc('xtick',labelsize=20,direction='in',top=True)
plt.rc('ytick',labelsize=20,direction='in',right=True)
plt.rc('legend',fontsize='xx-large',framealpha=1.0)
```

```python
# read in data from both files for folded 
# observational and model data
data_names = ['fold', 'RV', 'RV_err']
data = np.genfromtxt('ELCdataRV1.fold',names=data_names,unpack=False)
model_names = ['model_fold','model_RV']
model = np.genfromtxt('star1.RV',names=model_names,unpack=False)

# plot it please
plt.plot(model['model_fold'],model['model_RV']+0.1,c='black',lw=3)

my_data = (data['fold'] > 0.70) & (data['fold'] < 0.72)

plt.errorbar(data['fold'][~my_data],data['RV'][~my_data],yerr=data['RV_err'][~my_data],fmt='ro',alpha=0.75)
plt.errorbar(data['fold'][my_data],data['RV'][my_data],yerr=data['RV_err'][my_data],fmt='bo',alpha=0.75)

plt.grid()
plt.xlim(model['model_fold'][0],model['model_fold'][-1])
plt.xlim(0,1)
plt.ylim(data['RV'].min()-0.3,data['RV'].max()+0.3)
plt.xlabel('Phase')
plt.ylabel('RV (km/s)')
plt.savefig('56UMa_RV_Curve_combined.png')
```
